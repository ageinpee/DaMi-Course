```python
from sklearn.datasets.samples_generator import make_circles
from sklearn.datasets.samples_generator import make_moons
from sklearn.datasets.samples_generator import make_swiss_roll
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Specify number of samples, some noise is added
no_samples = 1500
noise_factor=.05

# Creates two circles, try to fit them with kmeans
circles = make_circles(n_samples=no_samples, factor=.5, noise=noise_factor)

kmeans = KMeans(n_clusters=3, random_state = 0)
kmeans.fit(circles[0])

# Plot the results
plt.scatter(circles[0][:, 0], circles[0][:, 1],c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 50, c = 'black')
```

```python
# Creates two half moons and try to fit them with kmeans
noisy_moons = make_moons(n_samples=no_samples, noise=noise_factor)

X = noisy_moons[0]

kmeans = KMeans(n_clusters=2, random_state = 0)
kmeans.fit(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1],c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 50, c = 'black')
```

```python
# Creates the popular 'swiss roll' data distribution and try to fit them with kmeans
swiss = make_swiss_roll(no_samples, noise=noise_factor)

X = swiss[0]

kmeans = KMeans(n_clusters=3, random_state = 0)
kmeans.fit(X)

# Plot the resultsfig = plt.figure()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2],c = kmeans.labels_)
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s = 50, c = 'black')
plt.show()
```

```python
# Try out the datasets with agglomerative clustering and different linkages (single, average, complete)

from sklearn.cluster import AgglomerativeClustering

X = circles[0]

agglo = AgglomerativeClustering(n_clusters=2, linkage="single")
agglo.fit(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1],c = agglo.labels_)
```

```python
X = noisy_moons[0]

agglo = AgglomerativeClustering(n_clusters=2, linkage="single")
agglo.fit(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1],c = agglo.labels_)
```

```python
X = swiss[0]

agglo = AgglomerativeClustering(n_clusters=6, linkage="complete")
agglo.fit(X)

# Plot the resultsfig = plt.figure()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2],c = agglo.labels_)
plt.show()
```

```python
# Try out the datasets with the DBSCAN algorithm

from sklearn.cluster import DBSCAN

X = circles[0]

dbscan = DBSCAN(eps=0.099)
dbscan.fit(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1],c = dbscan.labels_)
```

```python
X = noisy_moons[0]

dbscan = DBSCAN(eps=0.099)
dbscan.fit(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1],c = dbscan.labels_)
```

```python
X = swiss[0]

dbscan = DBSCAN(eps=0.099)
dbscan.fit(X)

# Plot the results
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2],c = dbscan.labels_)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
