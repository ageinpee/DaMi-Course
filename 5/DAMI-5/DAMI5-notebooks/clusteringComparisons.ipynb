```python
from sklearn.datasets.samples_generator import make_circles
from sklearn.datasets.samples_generator import make_moons
from sklearn.datasets.samples_generator import make_swiss_roll
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Specify number of samples, some noise is added
no_samples = 1500
noise_factor=.05

# Creates two circles, try to fit them with kmeans
circles = make_circles(n_samples=no_samples, factor=.5, noise=noise_factor)

X = circles[0]

kmeans = KMeans(n_clusters=3, random_state = 0)
kmeans.fit(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1],c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 50, c = 'black')
```

```python
# Creates two half moons and try to fit them with kmeans
noisy_moons = make_moons(n_samples=no_samples, noise=noise_factor)

X = noisy_moons[0]

kmeans = KMeans(n_clusters=2, random_state = 0)
kmeans.fit(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1],c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 50, c = 'black')
```

```python
from mpl_toolkits.mplot3d.axes3d import get_test_data
# Creates the popular 'swiss roll' data distribution and try to fit them with kmeans
swiss = make_swiss_roll(no_samples, noise=noise_factor)

X = swiss[0]

kmeans = KMeans(n_clusters=3, random_state = 0)
kmeans.fit(X)


# Plot the resultsfig = plt.figure()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2],c = kmeans.labels_)
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 2], s = 50, c = 'black')
plt.show()
```

```python
# Try out the datasets with agglomerative clustering and different linkages (single, average, complete)

from sklearn.cluster import AgglomerativeClustering

X = circles[0]

agglo = AgglomerativeClustering(n_clusters=2, linkage="single")
agglo.fit(X)
plt.scatter(X[:, 0], X[:, 1],c = agglo.labels_)
```

```python
agglo = AgglomerativeClustering(n_clusters=2, linkage="average")
agglo.fit(X)
plt.scatter(X[:, 0], X[:, 1],c = agglo.labels_)
```

```python
agglo = AgglomerativeClustering(n_clusters=2, linkage="complete")
agglo.fit(X)
plt.scatter(X[:, 0], X[:, 1],c = agglo.labels_)
```
